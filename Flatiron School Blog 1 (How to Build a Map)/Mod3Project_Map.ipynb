```python
poplist = resp.json()
pop_df = pd.DataFrame(poplist[1:], columns = poplist[0])
pop_df['FIPS'] = pop_df.state + pop_df.county
pop_df.POP = pop_df.POP.astype('int')
del pop_df['state']
del pop_df['county']
```

## Add in Rural Binning and merge onto pop_df

```python
rural_lookup = pd.read_excel('County_Rural_Lookup.xlsx', skiprows = 3)
```

```python
rural_lookup.drop(rural_lookup.tail(6).index,inplace=True)
```

```python
del rural_lookup['Note']
del rural_lookup['Unnamed: 8']
```

```python
rural_lookup.columns = ['FIPS', 'State', 'County_State', 'Total_pop', 'Urban_pop', 'Rural_pop', 'Percent_Rural']
```

```python
pop_df1 = pop_df.merge(rural_lookup[['FIPS', 'County_State', 'Percent_Rural']], on = 'FIPS', how = 'inner')
```

```python
def rural_marker(percent):
    marker = None
    if percent <= 30:
        marker = 'Urban'
    elif percent <= 70:
        marker = 'Mix/Suburban'
    else:
        marker = 'Rural'
    return marker
```

```python
pop_df1['pop_category'] = pop_df1.Percent_Rural.map(lambda x: rural_marker(x))
```

## Create FIPS Columns in map df

```python
map_df_CONUS['FIPS'] = map_df_CONUS.STATEFP + map_df_CONUS.COUNTYFP
```

## Merge together map and population dataframe

```python
merged_df = map_df_CONUS.merge(pop_df1, on = 'FIPS', how = 'left')
```

```python
merged_df.info()
```

## Mapping

```python
geoplot.choropleth(
    merged_df, hue = merged_df.pop_category,
    cmap='RdYlBu', figsize=(20, 20), legend = True)
fig = plt.gcf()
plt.show()
```
